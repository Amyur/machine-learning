# Extract the data

```python
import bs4
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from nltk.corpus import wordnet
from spellchecker import SpellChecker
from nltk.text import Text 
```

```python
responses = []
for i in range(10, 710, 5):
    response = requests.get("https://www.tripadvisor.co/Hotel_Review-g297473-d7213603-Reviews-or{}-Hotel_Four_Points_By_Sheraton_Barranquilla-Barranquilla_Atlantico_Department.html#REVIEWS".format(i))
    responses.append(response)
```

```python
soups = []
for i in range(len(responses)):
    soup = bs4.BeautifulSoup(responses[i].text, "html.parser")
    soups.append(soup)
```

# Analyse the data

```python
import nltk
import re
from nltk.corpus import stopwords
import pandas as pd
```

```python
df = pd.read_csv("reviews.csv")
```

```python
df.columns = ["idx", "review"]
```

```python
stoplist = stopwords.words("spanish")

def cleaning(text):
    text = (text.apply(lambda text: re.sub(r"[¡!\#$%&\(\)-=\\?¨\[_:,;.\]¡]","", text))
                .map(lambda text: re.sub(r"[0-9]","", text))
                .map(lambda text: text.lower())
                .map(lambda text: re.sub(r"\s+"," ", text))
                .map(lambda text: " ".join([i for i in text.split() if (i not in stoplist) and (len(i) > 2)])))
    return(text)
```

```python
text = cleaning(df.review)
```

```python
new_text = []
for sentence in text:
    if len(nltk.word_tokenize(sentence))<2:
        pass
    else:
        new_text.append(sentence)
```

```python
text = pd.Series(new_text)
```

```python
text_together = text.str.cat(sep=' ')
```

## Word-Cloud

```python
text_together = text_together.replace("hotel", "")
text_together = text_together.replace("barranquilla", "")
text_together = text_together.replace("habitaciones", "habitación")
text_together = text_together.replace("buen", "bueno")
text_together = text_together.replace("buenoa", "bueno")
text_together = text_together.replace("four", "")
text_together = text_together.replace("points", "")
text_together = text_together.replace("service", "")
text_together = text_together.replace("check", "")
text_together = text_together.replace("out", "")
text_together = text_together.replace("sheraton", "")
text_together = text_together.replace("room", "")
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=60, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()
```

```python
show_wordcloud(text_together)
```

## Topic modeling

```python
from gensim import corpora, models, similarities
from itertools import chain
import nltk
from nltk.corpus import stopwords
from operator import itemgetter
import re
from nltk.tokenize import word_tokenize
```

```python
texto = [[word for word in word_tokenize(text_together)]]
dictionary = corpora.Dictionary(texto)
corpus = [dictionary.doc2bow(text) for text in texto]
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
lsi = models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=100)
#print(lsi.print_topics(20))
n_topics=1
lda = models.LdaModel(corpus_tfidf, id2word=dictionary, num_topics=n_topics)
```

```python
for i in range(0, n_topics):
    temp = lda.show_topic(i, 10)
    terms = []
    for term in temp:
        terms.append(term)
    print ("Top 10 terms for topic #" + str(i) + ": "+ ", ".join([j[0] for j in terms]))
```

```python
for idx, topic in lda.show_topics(formatted=False, num_words= 10):
    print('Topic: {} \nWords: {}'.format(idx, [w[0] for w in topic]))
```

## Summaryse

```python
import nltk
results=[]

for sentence in text:
    
    no_of_tokens = len(nltk.word_tokenize(sentence))
    #print no_of_toekns
    
    # Let's do POS tagging
    tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
    
    # Count the no of Nouns in the sentence
    no_of_nouns = len([word for word, pos in tagged if pos in ["NN","NNP"] ])
     
    #Use NER to tag the named entities.
    ners = nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sentence)), binary=False)
    
    no_of_ners = len([chunk for chunk in ners if hasattr(chunk, 'label')])
    
    score = (no_of_ners+no_of_nouns)/float(no_of_tokens) #Fracción de tokens siendo entidades comparado con tokens normales
    
    results.append((no_of_tokens, no_of_ners, no_of_nouns, score, sentence))
```

```python
for sent in sorted(results[0:5], key=lambda x: x[3], reverse=True):
    print("Comentario: " + sent[4])
    print(" ")
```
